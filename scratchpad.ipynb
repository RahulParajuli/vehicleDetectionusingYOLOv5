```
import torch
from matplotlib import pyplot as plt
import numpy as np
```

```
model = torch.hub.load('ultralytics/yolov5','yolov5s')
```

```
img = 'https://ultralytics.com/images/zidane.jpg'
```

```
result = model(img)
result.print()
```

```
%matplotlib inline
plt.imshow(np.squeeze(result.render()))
plt.show()
```
